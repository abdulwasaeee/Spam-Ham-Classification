```python
import nltk
import numpy as np
import pandas as pd
import re
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('wordnet')
le=nltk.WordNetLemmatizer()
nltk.download('stopwords')
```

```python
msgs=pd.read_csv('spam.txt',sep='\t',names=['label','message'])
msgs.head()
```

```python
corpus=[]
for i in range(len(msgs)):
  review=re.sub('[^a-zA-z]',' ',msgs['message'][i])
  review=review.lower()
  review=review.split()
  review=[le.lemmatize(word) for word in review]
  review=" ".join(review)
  corpus.append(review)
```

```python
for i in corpus:
  print(i)
```

```python
[[i,j,k] for i,j,k in zip(list(map(len,corpus)),corpus, msgs['message']) if i<1]
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
nltk.download('punkt_tab')
```

```python
words=[]
for i in corpus:
  token=sent_tokenize(i)
  for j in token:
    words.append(simple_preprocess(j))
```

```python
words
```

```python
from gensim.models import Word2Vec
```

```python
model=Word2Vec(words)
```

```python
model.wv.index_to_key
```

```python
def aw2v(doc):
    vectors = [model.wv[word] for word in doc if word in model.wv.index_to_key]
    if not vectors:  # if no word embeddings found
        return np.zeros(model.vector_size)
    return np.mean(vectors, axis=0)
```

```python
!pip install tqdm
```

```python
from tqdm import tqdm

x = [aw2v(doc) for doc in tqdm(words)]
x2 = np.array(x)  # Now works: shape = (5569, vector_size)
```

```python
y=msgs[list(map(lambda x: len(x)>0,corpus))]
y=y['label'].map({'ham':0,'spam':1}).values
y
```

```python
df=pd.DataFrame(x2)
df['output']=y
df.columns = df.columns.astype(str) # Convert column names to strings
df.head()
```

```python
df.dropna(inplace=True)
```

```python
x=df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf= RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
pred=rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_test,pred)
```
